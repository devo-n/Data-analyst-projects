# Automatic File Sorter

```python
import os, shutil
```

```python
path = r"C:/Users/dyepa/Pictures/Feedback/"
```

```python
file_name = os.listdir(path)
```

```python
folder_names = ['Excel Workbook', 'image files']

for loop in range(0,2):
    if not os.path.exists(path + folder_names[loop]):
        print(path + folder_names[loop])
        os.makedirs((path + folder_names[loop]))

for file in file_name:
    if ".jpg" in file and not os.path.exists(path + "image files/" + file):
        shutil.move(path + file, path + "image files/" + file)
    if ".png" in file and not os.path.exists(path + "image files/" + file):
        shutil.move(path + file, path + "image files/" + file)
    if ".xlsx" in file and not os.path.exists(path + "Excel Workbook/" + file):
        shutil.move(path + file, path + "Excel Workbook/" + file)
```
